```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
import seaborn as sns
sns.set()
```

```python
def self_pointer():
    i = 0
    selfpoint = 0
    while i < 14:
        pfrom = random.choice(range(10))

        pto = random.choice(range(10))

        #print(str(pfrom) + " points to " + str(pto))

    return [pfrom, pto]
```

```python
j = 0
counter = []
while j < 5000:
    counter.append(self_pointer())
    j+= 1
```

```python
def self_pointer_big():
    i = 0
    selfpoint = 0
    while i < 519:
        pfrom = random.choice(range(424))

        pto = random.choice(range(424))

        #print(str(pfrom) + " points to " + str(pto))

        if pfrom == pto:
            selfpoint+= 1

        i+= 1

    return selfpoint
```

```python
#math bio stuff
plt.rcParams.update({'font.size': 16})

y = np.linspace(1.00000001,33.7,5000)
r = np.zeros(5000)
k = np.zeros(5000)

for i in range(len(y)):
    r[i] = 2*y[i]**3 / ((1+(y[i]**2))**2)
    
    k[i] = 2*y[i]**3/((y[i]**2)-1)
    
plt.plot(k,r)
plt.xlim(0,40)
plt.ylim(0,0.8)
plt.xlabel('nondimensionalized growth rate')
plt.ylabel('nondimensionalized carrying capacity')
plt.title('2-dimensional Bifurcation Diagram')

plt.show()
```

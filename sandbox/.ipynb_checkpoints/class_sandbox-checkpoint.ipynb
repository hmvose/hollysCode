```python
# initialize our sliders
wt_slider = pn.widgets.FloatSlider(name="Z_wt", start=0.5, end=4, step=0.5, value=1)

# define the plotting function, such that it depends on the values of the slider
@pn.depends(wt_slider.param.value)
def rate_plot_NAR(Z_wt, eta_o = 0.02, delta = 0.17, K_y = 0.4):
    
    # x and y values to plot
    M = 1.8*Z_wt
    Z = np.linspace(0, 1.5, 100)
    g1 = -(eta_o*Z)/(1-(Z/M)) + delta*(Z*0.1)/(K_y+0.1)
    g2 = -(eta_o*Z)/(1-(Z/M)) + delta*(Z*0.6)/(K_y+0.6)
    g3 = -(eta_o*Z)/(1-(Z/M)) + delta*(Z*5)/(K_y+5)
    
    # specify the list of plots we want
    list_of_curves = [
    hv.Curve(data=(Z, g), kdims=["Z"], vdims=["dg/dZ"], label="growth rate").opts(color="black"),
    ]
    
    # overlay all the plots, fix ylim so it doesn't change with the sliders
    return hv.Overlay(list_of_curves).opts(ylim=(0, 3), show_grid=True, width=400, height=300)

# layout the sliders and the plot
pn.Column(wt_slider, rate_plot_NAR, width=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
import seaborn as sns
sns.set()
```

```python
def self_pointer():
    i = 0
    selfpoint = 0
    while i < 14:
        pfrom = random.choice(range(10))

        pto = random.choice(range(10))

        #print(str(pfrom) + " points to " + str(pto))

    return [pfrom, pto]
```

```python
j = 0
counter = []
while j < 5000:
    counter.append(self_pointer())
    j+= 1
```

```python
def self_pointer_big():
    i = 0
    selfpoint = 0
    while i < 519:
        pfrom = random.choice(range(424))

        pto = random.choice(range(424))

        #print(str(pfrom) + " points to " + str(pto))

        if pfrom == pto:
            selfpoint+= 1

        i+= 1

    return selfpoint
```

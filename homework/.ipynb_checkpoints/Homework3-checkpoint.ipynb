`[ [35, 234], [264, 67], . . . , [89, 122] ]` 

```python
def network_gen(arrows, nodes):
    i = 0
    selfpoint = 0
    while i < arrows:
        pfrom = random.choice(range(nodes))

        pto = random.choice(range(nodes))

        #print(str(pfrom) + " points to " + str(pto))

        i+= 1

    return [pfrom, pto]
```

Using this lists of lists you have saved, write a function that will count the number of feedforward loops in this ER network. The approach here is up to you although you can ask me for suggestions! Make sure to document / comment your code clearly to explain your approach. Again compare your results to the real network shown in Figure 3.3, where there are a total of 42 feedforward loops.

$\text{The basic idea behind my code is to only focus on the tuples that are capable of being in a FFL situation.}$
$\text{Specifically, we can narrow down which tuples could form an FFL, and then check if their missing link exists.}$
$\text{In order to do this, we will manipulate the network list to cherry pick the tuples we care about.}$

```python
#HELPER FUNCTION 1: The Separator Function creates a unique list of all of the numbers that appear as a node that was pointed
#to, and a node that is being pointed from. This is the starting point for trying to find specific tuples.
#For further reference, I am considering a [start, end] model here.

def separator(network):
    #initialization of lists
    start_list = []
    end_list = []
    
    #separate every tuple into their starting nodes and ending nodes
    for num in network:
        start_list.append(num[0])
        end_list.append(num[1])
        
    #the lists are then sorted and converted to sets to ensure that each number is unique
    start_list.sort()
    ustart = list(set(start_list))
    end_list.sort()
    uend = list(set(end_list))
    
    #the unique start and unique end lists are returned for further use    
    return ustart, uend
        
```

```python
#HELPER FUNCTION 2: The Group Finder function takes in a tuple and searches for all instances of it within a
#provided network. It also contains information about how many of the requested tuples it has found, and whether
#or not the network has any specific tuples left (in case of duplicate FFLs).

def group_finder(find_group, network_copy, counter):
    
    #checking the desired tuple to see if it exists in the provided network
    for num in network_copy:
        
        #if the tuple exists, we can increment our counter for the pattern and remove it so it won't loop forever
        if find_group == num:
            counter += 1
            network_copy.remove(find_group)        #remove will only take out the first instance of the tuple
            only_group = False                     #because there may be duplicates this function must occur twice 
        else:                                      #in order to properly cease
            only_group = True                      #if no tuple is found, then the function can quit
        
    return only_group, counter
```

```python
#THE BIG FUNCTION: The FFL Finder takes in a network and mutilates a copy in order to find the number of patterns
#present. It utilizes both helper functions previously mentioned.
#Code created by Holly Vose

def ffl_finder(network):
    
    #STEP 1: We make a copy of the network we can edit and find all unique start and end values
    network_copy = network.copy()
    ustart, uend = separator(network_copy)
    
    #STEP 2: Compare the start and end list to see which numbers are present in both. If there is an overlap, then
    #there is a possibility of a FFL motif
    overlap_nums = list(set(ustart).intersection(uend))
    
    #STEP 2.5: Initialize new lists for editing
    tops = []
    bottoms = []

    #STEP 3: Now, we must extract all of the tuples that have one instance of the overlapped numbers. This model is
    #different from the separator function, where the connection is [start, top] and [bottom, end]. Top and bottom
    #in this case must be equal by virtue of the overlap_nums list. 
    for num in overlap_nums:
        for group in network_copy:
            if group[0] != group[1]:               #this ensures we don't have a self-pointing network, since FFL
                if group[0] == num:                #motifs must have at least three nodes to be considered an FFL
                    bottoms.append(group)
                if group[1] == num:
                    tops.append(group)
                    
    #STEP 3.5: Once the possible FFL segments are sorted, a new counter is initialized to tell us how many FFLs are
    #present in the network.
    ffl_counter = 0
    
    #STEP 4: Now, we must match the tops to their respective bottom parts. For example, if the tuple [1,2] was in
    #one list, and the tuple [2,3] was in the other array, the program recognizes the overlapping 2. It then makes
    #a new tuple by removing the overlap and combining what's left: the tuple ([1,3]) that would complete the FFL.
    for top in tops:
        for bottom in bottoms: 
            only_group = False                     #recall that the group finder can only remove one instance of a
            if top[1] == bottom[0]:                #tuple at a time, so we must introduce a boolean checkpoint
                find_group = [top[0], bottom[1]]
                
    #STEP 5: The Group Finder function is called to see if the missing tuple exists in the network. If it does, the
    #counter is updated and the function continues. 
                while only_group is False:
                    group_status = group_finder(find_group, network_copy, ffl_counter)
                    only_group = group_status[0]
                    ffl_counter = group_status[1]

    #STEP 6: Return the final count of FFLs in the network. This way, the function does not have to brute force
    #search every single tuple for a pattern, but rather can assemble a list of 'possible' patterns and then 
    #search for the missing link.
    return ffl_counter
```

## b)

Next use the differential equations below to numerically integrate the behavior for $Y$ and $Z$ in response to the array of $X$ values you set up in part a).

$$ \frac{dY}{dt} = \beta_Y \frac{X^n}{K_{XY}^n + X^n} - \alpha_Y Y $$

$$ \frac{dZ}{dt} = \beta_Z \frac{X^n}{K_{XZ}^n + X^n} \frac{Y^n}{K_{YZ}^n + Y^n} - \alpha_Z Z $$

Euler integration is fine here, as long as you select an appropriate value of $\Delta t$. However as a challenge, you can also try to integrate this with `scipy`, although this is quite a bit trickier since there is not only a time array, but also the array of $X$ values upon which $Y$ and $Z$ depend. 

For an example of how to integrate more than one differential equation at a time, you can check out my previous lecture notes here:

https://sbeeler.github.io/BIOL301_Fall2022/04_scipy_integration.html

These notes address how you might integrate $Y$ and $Z$ concurrently, but they don't address how you would incorporate the dependence on the $X$ array. So if you want to take this approach, you will still have to figure that piece out.

```python
#creating the time arrays
Yi = np.zeros(150)
Zi = np.zeros(150)

#n init
n = 2

#production
betaYi = 5
betaZi = 7

#degradation
alphaYi = 1
alphaZi = 2

#binding affinities
K_xyi = 3
K_yzi = 7
K_xzi = 10

#dt step
dti = 0.08 #minutes

for t in range(len(Yi)-1):
    dYi = (betaYi*Xffl[t]**n)/(K_xyi**n+Xffl[t]**n) - alphaYi*Yi[t]
    Yi[t+1] = Yi[t] + dYi*dti
    
    dZi = (betaZi*Xffl[t]**n)/(K_xzi**n+Xffl[t]**n)*(betaZi*Yi[t]**n)/(K_yzi**n+Yi[t]**n) - alphaZi*Zi[t]
    Zi[t+1] = Zi[t] + dZi*dti
    
t = np.arange(len(Yi))*dti
plt.plot(t, Xffl)
plt.plot(t,Yi)
plt.plot(t,Zi)
plt.xlabel("time (mins)")
plt.ylabel("copy number of X per cell")
plt.legend(["X production", "Y production", "Z production"])
    
    
```

$$ \frac{dY}{dt} = \beta_Y \frac{X^n}{K_{XY}^n + X^n} - \alpha_Y Y $$

$$ \frac{dZ}{dt} = \beta_Z \frac{X^n}{K_{XZ}^n + X^n}\frac{K_{YZ}}{K_{YZ}^n + Y^n} - \alpha_Z Z $$

$\text{The primary difference here from the C1-FFL is the introduction of the repression in the second hill equation.}$

$\text{The Y component of the dZ equation is actively repressing function, while X is still activating.}$
$\text{This leads to the activating/repressing combination from both X and Y on Z.}$

## b)

Conduct your integration and display your results below.

```python
#creating the time arrays
X = np.zeros(1000)
Y = np.zeros(1000)
Z = np.zeros(1000)
R = np.zeros(1000)
  
#n init
n = 3 

#simple regulation init
betaR = 4.2 #production
alphaR = 1 #degradation
K = 5 

#Y Init
betaY = 10 #production
alphaY = 1 #degradation

#Z init
betaZ = 170 #production
alphaZ = 10 #degradation


#binding affinities
K_yz = 10 #Y binds to Z
K_xy = 1 #X binds to Y 
K_xz = 10 #X binds to Z


dt = 0.01 # minutes

#simple regulation
for t in range(100, len(R)-1):
    dR = betaR - alphaR * R[t]
    R[t+1] = R[t] + dR*dt

#X over time
X[100:1000] = 10 #steady state production


#Plotting Y and Z over time
for t in range(len(Y)-1):
    dY =  betaY*(X[t]**n) / (K_xy**n + X[t]**n) - alphaY*Y[t]
    Y[t+1] = Y[t] + dY*dt
      
    dZ = betaZ*((X[t]**n) / (K_xz**n + X[t]**n)) * ((K_yz**n) / (K_yz**n + Y[t]**n)) - alphaZ*Z[t]
    Z[t+1] = Z[t] + dZ*dt
    
t = np.arange(len(Z))*dt 
plt.plot(t, R, label = "Simple Regulation", color='k')
plt.plot(t, X, label = "X",color='b')
plt.plot(t, Y, label = "Y", color = 'r')
plt.plot(t, Z, label = "Z", color = 'g')

plt.xlabel("time (mins)")
plt.ylabel("copy number of X per cell")
plt.legend(loc='best')
 
```

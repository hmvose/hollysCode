# Problem 2: Adding noise to get oscillations

In this problem, we will explore the claim on pg. 101 of Alon that noise can help maintain oscillations. We will be basing this off of the system shown in Figure 6.5.

- (1 pt.) Write out the differential equations for this system, specifying the Hill equations fully.

$\textbf{X EQUATION: } \frac{dX}{dt} = \beta_X \frac{K_X^n}{K_X^n+Y^n} - \alpha X$

$\textbf{Y EQUATION: } \frac{dY}{dt} = \beta_Y \frac{X^n}{K_Y^n+X^n} - \alpha Y$

- (1 pt.) Numerically integrate (like with Euler) and plot the results of X and Y over time. You should see damped oscillations. Use the discussion on pg. 99-100 to guide your parameter choices to make sure your system isn't too overdamped. 

- (1 pt.) Now add noise into the system when you are doing the numerical integration. Instead of having a definite amount of $dX$ for example, now allow $dX$ to vary by a multiplicative factor of `noise` either up or down. You should to use the `random` module to help you randomly select a value of $dX$ in this way. 

- (1 pt.) Replot your integration, this time with the noise incorporated. With the right parameters you should now see sustained oscillations. Depending on the value of your other parameters, the noise might need to be unreasonably high (like 10) to get oscillations to occur, but it still demonstrates the point that noise can facilitate oscillations. 


```python
Kx = 0.3
Ky = 0.3
beta_x = 0.5
beta_y = 0.4
n = 7
alpha_x = 0.3
alpha_y = 0.3
```

```python
#system with noise
#system without noise
dt = 0.1
X = np.zeros(300)
Y = np.zeros(300)

for i in range(len(X)-1):
    dX = beta_x * (Kx**n)/(Kx**n + Y[i]**n) - alpha_x*X[i]
    X[i+1] = X[i] + dX*dt
    
    dY = beta_y * (X[i]**n)/(X[i]**n + Ky**n) - alpha_y*Y[i]
    Y[i+1] = Y[i] + dY*dt
    
t = np.arange(len(X))*dt

plt.plot(t,X)
plt.plot(t,Y)
plt.legend(["X Production","Y Production"])
```

```python
#system without noise
dt = 0.1
X_r = np.zeros(300)
Y_r = np.zeros(300)

for i in range(len(X_r)-1):
    noise_x = random.randint(-2,2)
    noise_y = random.randint(-2,2)
    
    dX = beta_x * (Kx**n)/(Kx**n + Y_r[i]**n) - alpha_x*X_r[i]
    X_r[i+1] = X_r[i] + dX*dt*noise_x
    
    dY = beta_y * (X_r[i]**n)/(X_r[i]**n + Ky**n) - alpha_y*Y_r[i]
    Y_r[i+1] = Y_r[i] + dY*dt*noise_y
    
t = np.arange(len(X_r))*dt

plt.plot(t,X_r)
plt.plot(t,Y_r)
plt.legend(["X production","Y production"])
    
```

```python
#making the four plots
x = np.linspace(0,4,200)
y = np.linspace(0,4,200)
beta_low = 5
beta_one = 8
beta_two = 10
beta_large = 21
beta_y = 10

K = 1
n = 3
alpha = 5


X_prod_low = beta_low * K**n / (K**n + x**n)
X_prod_one = beta_one * K**n / (K**n + x**n)
X_prod_two = beta_two * K**n / (K**n + x**n)
X_prod_large = beta_large * K**n / (K**n + x**n)
Y_prod = beta_y * K**n / (K**n + y**n)
    
null_x_l = X_prod_low / alpha
null_x_o = X_prod_one / alpha
null_x_t = X_prod_two / alpha
null_x_g = X_prod_large / alpha
null_y = Y_prod / alpha

figure, axis = plt.subplots(1,4, figsize = (16,5))
figure.suptitle("Varying beta values")

axis[0].plot(null_x_l, y)
axis[0].plot(x, null_y)
axis[0].legend(["null x","null y"])
axis[0].set_title("low beta (5)")
axis[0].set_ylabel("y num")
axis[0].set_xlabel("x num")

axis[1].plot(null_x_o, y)
axis[1].plot(x, null_y)
axis[1].legend(["null x","null y"])
axis[1].set_title("lower threshold beta (8)")
axis[1].set_xlabel("x num")

axis[2].plot(null_x_t, y)
axis[2].plot(x, null_y)
axis[2].legend(["null x","null y"])
axis[2].set_title("upper threshold beta (10)")
axis[2].set_xlabel("x num")

axis[3].plot(null_x_g, y)
axis[3].plot(x, null_y)
axis[3].legend(["null x","null y"])
axis[3].set_title("high beta (21)")
axis[3].set_xlabel("x num")

```

```python
def beta_input(upper):
    change = upper / 100
    k_time = np.zeros(200)
    i = 0

    while i < 100:
        k_time[i] += change
        k_time[i+1] = k_time[i]
        i += 1

    while i < 199:
        if k_time[i] == 0:
            break;
        k_time[i] -= change
        k_time[i+1] = k_time[i]
        i += 1
        
    k_time[199] = 0
    
    return k_time
```

```python
new_beta = beta_input(20)
plt.plot(new_beta)
plt.title("input parameter")
plt.xlabel("time")
plt.ylabel("concentration")
```

```python
#EULER TIME
x = np.zeros(200)
y = np.zeros(200)
dt = 0.1
K = 4
alpha_x = 1
alpha_y = 0.5
beta_y = 4
n = 2
for i in range(len(x)-1):
    
    dX = new_beta[i] * K**n / (K**n + y[i]**n) - alpha_x*x[i]
    x[i+1] = x[i] + dX*dt
    #print(x[i+1])
    
    dY = beta_y * K**n / (K**n + x[i]**n) - alpha_y*y[i]
    y[i+1] = y[i] + dY*dt

t = np.arange(len(x))*dt

plt.plot(t, x)
plt.plot(t, y)
plt.title("hysteresis")
plt.xlabel("time")
plt.ylabel("concentration")
#this homework needs more time on the redo, i just didn't have the time to fully troubleshoot this week.
```

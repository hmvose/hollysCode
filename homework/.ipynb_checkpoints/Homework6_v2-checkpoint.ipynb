```python
X = np.ones(800)

X[200:400] = 2
X[400:600] = 4
X[600:800] = 8

plt.plot(X)
plt.xlabel("time")
plt.ylabel("X");
```

$\textbf{Requirements for exact adaptation:}$

$\text{When at steady state } (\frac{dY}{dt} = 0, \frac{dZ}{dt} = 0) \text{ the dependence of z is not reliant on any variables like X or Y.}$

$\textbf{Requirements for fold change detection:}$

$\text{The equation can be rewritten in a generalized form in terms of some dimensionless } \tilde{Y} = \frac{Y}{X_o}$
$\text{ and the fold change variable, } F = \frac{X}{X_o} \text{ with no other dependencies on extra variables.}$

$\textbf{System 1:}$

$$\frac{dY}{dt} = x^2 - y$$

$$\frac{dZ}{dt} = \frac{x}{y} - z$$

```python
# parameters
dt = 0.1

# initilize Y and Z arrays to integrate
Y = np.zeros(len(X))
Z = np.zeros(len(X))
Y[0] = 1

# integration of Y and Z over time
for t in range(len(X)-1):
    dY = X[t]*X[t] - Y[t]
    Y[t+1] = Y[t] + dY*dt
    
    dZ = X[t]/Y[t] - Z[t]
    Z[t+1] = Z[t] + dZ*dt  
    
t = np.arange(len(X))*dt 

# plotting
fig, ax = plt.subplots(3,1, figsize=(5,7))
ax[0].plot(t,X)
ax[0].set_ylabel("X")

ax[1].plot(t,Y)
ax[1].set_ylabel("Y")

ax[2].plot(t,Z)
ax[2].set_ylabel("Z")
ax[2].set_xlabel("time");
```

$\text{Does this system show Exact Adaptation?} \textbf{ No.}$
$\text{We can clearly see in the Z graph that the system does not return to the same steady state after each change.}$

$\textbf{Proof that Exact Adaptation is not present:}$

$$ 0 = x^2 - y \text{ and } 0 = \frac{x}{y} - z$$

$$ y = x^2 \text{ and } z = \frac{x}{y}$$

$$\text{With substitution: } z = \frac{x}{x^2}$$

$$z = \frac{1}{x}$$

$\text{We can clearly see that z is dependent on x, and therefore does not display exact adaptation.}$

$\text{Does this system show Fold Change Detection?} \textbf{ No.}$
$\text{Because this system does not show exact adaptation, it is not possible for the system to show fold change detection,}$
$\text{as exact adaptation is required for FCD.}$

$\textbf{System 2:}$

$$\frac{dY}{dt} = (x - y)^2$$

$$\frac{dZ}{dt} = \frac{x}{y} - z$$

```python
# parameters
dt = 0.1

# initilize Y and Z arrays to integrate
Y = np.zeros(len(X))
Z = np.zeros(len(X))
Y[0] = 1

# integration of Y and Z over time
for t in range(len(X)-1):
    dY = (X[t] - Y[t])*(X[t]-Y[t])
    Y[t+1] = Y[t] + dY*dt
    
    dZ = X[t]/Y[t] - Z[t]
    Z[t+1] = Z[t] + dZ*dt  
    
t = np.arange(len(X))*dt 

# plotting
fig, ax = plt.subplots(3,1, figsize=(5,7))
ax[0].plot(t,X)
ax[0].set_ylabel("X")

ax[1].plot(t,Y)
ax[1].set_ylabel("Y")

ax[2].plot(t,Z)
ax[2].set_ylabel("Z")
ax[2].set_xlabel("time");
```

$\text{Does this system show Exact Adaptation?} \textbf{ Yes.}$
$\text{After each change in the system, we see that z returns to the same steady state each time.}$

$\textbf{Proof of Exact Adaptation:}$

$$0 = x^2 + 2xy + y^2 \text{ and } 0 = \frac{x}{y} - z$$

$$\text{By rearranging, we get } x = zy, \text{ which by substitution gives }z^2 y^2 - 2 z y^2 + y^2 = 0$$

$$y^2(z^2-2z+1) = 0. \text{which then becomes } y^2(z-1)(z-1) = 0$$

$$\text{In any case, we find } z = 1$$

$\text{Here, we see that z is not dependent on any variables, and therefore shows exact adaptation.}$

$\text{Does this system show Fold Change Detection?} \textbf{ No.}$
$\text{By looking at the graph, we see that the height of each pulse is impacted by the change of y. Therefore, it is not FCD.}$

$\textbf{Proof that FCD is not present:}$

$\text{The issue lies in the } \frac{dY}{dt} \text{ equation.}$

$$\frac{1}{X_o} * \frac{dY}{dt} = \frac{x^2}{x_o} - \frac{2xy}{x_o} + \frac{y^2}{x_o}$$

$$\frac{d\tilde{Y}}{dt} = Fx - 2Fy + \tilde{Y}y$$

$\text{Clearly we can see that x and y are still present in the final form of this equation, and there is no way to simplify}$
$\text{in just terms of F and }\tilde{Y}. \text{Therefore, FCD is not present.}$

$\textbf{System 3:}$

$$\frac{dY}{dt} = zy(z - z_o)$$

$$\frac{dZ}{dt} = \frac{x}{y} - z$$

```python
# parameters
dt = 0.1
z_o = 1

# initilize Y and Z arrays to integrate
Y = np.zeros(len(X))
Z = np.zeros(len(X))
Y[0] = 1

# integration of Y and Z over time
for t in range(len(X)-1):
    dY = Z[t]*Y[t]*(Z[t]-z_o)
    Y[t+1] = Y[t] + dY*dt
    
    dZ = X[t]/Y[t] - Z[t]
    Z[t+1] = Z[t] + dZ*dt  
    
t = np.arange(len(X))*dt 

# plotting
fig, ax = plt.subplots(3,1, figsize=(5,7))
ax[0].plot(t,X)
ax[0].set_ylabel("X")

ax[1].plot(t,Y)
ax[1].set_ylabel("Y")

ax[2].plot(t,Z)
ax[2].set_ylabel("Z")
ax[2].set_xlabel("time");
```

$\text{Does this system show Exact Adaptation?} \textbf{ Yes.}$
$\text{We see here that each pulse returns to the same steady state, showing exact adaptation.}$

$\textbf{Proof of Exact Adaptation:}$

$$0 = zy(z-z_o) \text{ and } 0 = \frac{x}{y} - z$$

$$zyz = zyz_o \text{ and } z = \frac{x}{y}$$

$$\text{Looking at the first equation, we see } z = z_o$$

$\text{Since } z_o \text{ is a constant parameter, this shows no dependence on any variables and therefore shows exact adaptation.}$

$\text{Does this system show Fold Change Detection?} \textbf{ Yes.}$
$\text{Although each pulse has a bit of noise associated with it, each one is exactly the same as the other pulses, showing FCD.}$

$\textbf{Proof that FCD is present:}$

$\text{First, look at } \frac{dY}{dt}$

$$\frac{1}{x_o} * \frac{dY}{dt} = \frac{zy(z-z_o)}{x_o}$$

$$\frac{d\tilde{Y}}{dt} = z\tilde{Y}(z-z_o)$$

$\text{Here, we see that } \frac{d\tilde{Y}}{dt} \text{ is only in terms of } \tilde{Y} \text{ and z. This shows FCD.}$

$\text{Next, look at } \frac{dZ}{dt}$

$$\frac{dZ}{dt} = \frac{x}{y} * \frac{x_o}{x_o} - z$$

$$\frac{dZ}{dt} = \frac{F}{\tilde{Y}} - z$$

$\text{Once again, we see that the only variables in this equation are F, }\tilde{Y},\text{ and z.}$

$\text{Since both equations are not dependent on x or y, we see FCD is present.}$


$\textbf{System 4:}$

$$\frac{dY}{dt} = x - y$$

$$\frac{dZ}{dt} = x - yz$$

```python
# parameters
dt = 0.1

# initilize Y and Z arrays to integrate
Y = np.zeros(len(X))
Z = np.zeros(len(X))
Y[0] = 1

# integration of Y and Z over time
for t in range(len(X)-1):
    dY = X[t] - Y[t]
    Y[t+1] = Y[t] + dY*dt
    
    dZ = X[t] - Y[t]*Z[t]
    Z[t+1] = Z[t] + dZ*dt  
    
t = np.arange(len(X))*dt 

# plotting
fig, ax = plt.subplots(3,1, figsize=(5,7))
ax[0].plot(t,X)
ax[0].set_ylabel("X")

ax[1].plot(t,Y)
ax[1].set_ylabel("Y")

ax[2].plot(t,Z)
ax[2].set_ylabel("Z")
ax[2].set_xlabel("time");
```

$\text{Does this system show Exact Adaptation?} \textbf{ Yes.}$
$\text{We see the return to the same steady state after each pulse, showing exact adaptation.}$

$\textbf{Proof of Exact Adaptation:}$

$$0 = x - y \text{ and } 0 = x - yz$$

$$x = y \text{ and } x = yz$$

$$\text{With substitution, we find } yz = y, \text{ which becomes } z = 1$$

$\text{Here, we see that z is not dependent on any variables, and therefore shows exact adaptation.}$

$\text{Does this system show Fold Change Detection?} \textbf{ No.}$
$\text{Each of the previous pulses is smaller than the next, and therefore does not show FCD.}$

$\textbf{Proof that FCD is not present:}$

$\text{The problem lies in } \frac{dZ}{dt}.$

$$\frac{dZ}{dt} = x * \frac{x_o}{x_o} - yz$$

$$\frac{dZ}{dt} = Fx_o - yz$$

$\text{Clearly, this shows that y is still present in the equation. But what happens if we multiply the whole equation by }\frac{1}{x_o}?$

$$\frac{dZ}{dt*x_o} = F - \tilde{Y}z$$

$\text{Now, our rate of change is divided by a random factor of } x_o. \text{ Therefore, we do not have FCD.}$

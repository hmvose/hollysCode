# Problem 2: Encoding "memory" in a genetic circuit 

Below we will explore the following circuit (from Figure 5.15), trying to recreate the results:



- (1 pt) Explain what is meant by "memory" in this system. To achieve memory, do you think $X$ and $Y$ are regulated with an AND gate or an OR gate? Why?

$\text{The memory of a system is where a system can be turned on and then can remain on, even in the absence of the original signal.}$

$\text{This is going to be regulated with an OR gate. This is because only one signal will be required to maintain the production,}$
$\text{even if it is not the original signal.}$

- (1 pt) Write down a system of differential equations for $X$ and $Y$. As a reminder, the production at an OR gate is encoded the following way, with an example of a gene $Z$ (not the one in our current circuit) being regulated by $X$ OR $Y$:

\begin{align}
\text{production of Z } &= \beta_Z \frac{x^{n_x} + y^{n_y} + x^{n_x}y^{n_y} }{(1 + x^{n_x})(1 + y^{n_y})}, \\[1em]
\text{where } x &= \frac{x}{k_{xz}} \text{ and } y = \frac{y}{k_{yz}} 
\end{align}

$\textbf{PRODUCTION OF X AND Y}$

\begin{align}
\text{production of X } &= \beta_X \frac{y^{n_y} + z^{n_z} + y^{n_y}z^{n_z} }{(1 + y^{n_y})(1 + z^{n_z})} - \alpha_X X, \\[1em]
\text{where } y &= \frac{Y}{k_{yx}} \text{ and } z = \frac{Z}{k_{zx}} 
\end{align}

\begin{align}
\text{production of Y } &= \beta_Y \frac{x^{n_x} + z^{n_z} + x^{n_x}z^{n_z} }{(1 + x^{n_x})(1 + z^{n_z})} -\alpha_Y Y, \\[1em]
\text{where } x &= \frac{X}{k_{xy}} \text{ and } z = \frac{Z}{k_{zy}} 
\end{align}

Note: To make the notation simpler, each $x$ is actually $x/k_{xz}$ and each $y$ is actually $y/k_{yz}$, where these two genes' amounts have been normalized by their binding coefficients to the $Z$ gene. As an aside, you may want to (re-)convince yourself that this equation does in fact encode an OR gate, where $Z$ will be "on" as long as either $X$ or $Y$ (or both) are present. 

- (0.5 pt) Make a "pulse" in $Z$ as the input to the circuit. Plot your value of $Z$ over time. 

- (1 pt) Integrate $X$ and $Y$ in response to $Z$. Plot the results of $X$, $Y$, and $Z$, either on one plot or in separate panels. 

- (0.5 pt) Explain what part (i.e. which arrows) of this circuit allows for "memory" to occur. 

```python
#setting the z values and making the pulse
z_mem = np.zeros(150)
z_mem[5:25] = 10

#plot the pulse
plt.plot(z_mem)
plt.xlabel("time")
plt.ylabel("value")
plt.title("Z Pulse")
```

```python
#alllllll the constants

#steady state values
beta_x = 10
beta_y = 15

#k for dX/dt
k_yx = 3
k_zx = 5

#k for dY/dt
k_xy = 1
k_zy = 4

#n values
n_x = 3
n_y = 2
n_z = 2

#degradation
alpha_X = 1
alpha_Y = 1
```

```python
#set the arrays to zero
xplot = np.zeros(150)
yplot = np.zeros(150)
dt = 0.1 # min

for i in range(len(xplot)-1):
    #new loop iteration means new values
    yx_dim = yplot[i] / k_yx
    xy_dim = xplot[i] / k_xy
    zx_dim = z_mem[i] / k_zx
    zy_dim = z_mem[i] / k_zy
    
    #dX integration by euler
    dX = beta_x * (yx_dim**n_y + zx_dim**n_z + (yx_dim**n_y)*(zx_dim**n_z)) / ((1 + yx_dim**n_y)*(1 + zx_dim**n_z)) - alpha_X*xplot[i]
    xplot[i+1] = xplot[i] + dX*dt
    
    #dY integration by euler
    dY = beta_y * (xy_dim**n_x + zy_dim**n_z + (xy_dim**n_x)*(zx_dim**n_z)) / ((1 + xy_dim**n_x)*(1 + zx_dim**n_z)) - alpha_Y*yplot[i]
    yplot[i+1] = yplot[i] + dY*dt
    
#plot everything
t = np.arange(len(xplot))*dt   
plt.plot(t, xplot)
plt.plot(t, yplot)
plt.plot(t, z_mem)
plt.legend(["dX", "dY", "Z pulse"])
plt.title("Memory of X and Y")
plt.xlabel("time")
plt.ylabel("production level")
```

$\text{The arrows present between X and Y are what allows this system to retain a memory of a previous signal.}$
$\text{The important factor to consider here is that the strength of X and Y against each other should be stronger}$
$\text{than Z's impact on them individually. This ensures that once Z is gone, the function will continue to maintain steady state.}$

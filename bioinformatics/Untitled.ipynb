```python
# -*- coding: utf-8 -*-
"""align.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1kHyoWNeqnCaspHDb5TELsCMa7aZbQdnm
"""

A = "MKE"
B = "MKLE"

GAP = -3
MATCH = 7
MISMATCH = -5

N = len(A)
M = len(B)

def max(A, B, C):
    if (A>=B and A>=C):
        return A
    elif (B>=A and B>=C):
        return B
    else:
        return C

def Tmax(A, B, C):
    if (A>=B and A>=C):
        return 'D'
    elif (B>=A and B>=C):
        return 'L'
    else:
        return 'U'

def m(p, q):
    if (p==q):
        return MATCH
    else:
        return MISMATCH

RA = ""
RB = ""
RM = ""

S = [[0 for _ in range(M+1)] for _ in range(N+1)]
T = [[0 for _ in range(M+1)] for _ in range(N+1)]

for i in range(N+1):
    S[i][0]=GAP*i
    T[i][0]='U'

for j in range(M+1):
    S[0][j]=GAP*j
    T[0][j]='L'

for i in range(1,N+1):
    for j in range(1,M+1):
        S[i][j]=max(S[i-1][j-1]+m(A[i-1],B[j-1]),S[i][j-1]+GAP,S[i-1][j]+GAP)
        T[i][j]=Tmax(S[i-1][j-1]+m(A[i-1],B[j-1]),S[i][j-1]+GAP,S[i-1][j]+GAP)

print("The score of the alignment is : "+str(S[N][M]))

i=N
j=M

while i!=0 or j!=0:
    if T[i][j]=='D':
        RA = A[i-1]+RA
        RB = B[j-1]+RB
        if A[i-1]==B[j-1]:
            RM = '|'+RM
        else:
            RM = "*"+RM
        i=i-1
        j=j-1
    elif T[i][j]=='L':
        RA = '-'+RA
        RB = B[j-1]+RB
        RM = ' '+RM
        j=j-1
    elif T[i][j]=='U':
        RA = A[i-1]+RA
        RB = '-'+RB
        RM = ' '+RM
        i=i-1

print(RA)
print(RM)
print(RB)
```

```python
inputFile = open("/content/network_098.txt","r")

graph = {}

for line in inputFile:
    info = line.split("\t")
    gene1 = info[0].strip()
    gene2 = info[1].strip()
    pcc = info[2].strip()
    if gene1 in graph:
        neighborList = graph[gene1]
        neighborList[gene2] = pcc
    else:
        graph[gene1] = {gene2:pcc}

    if gene2 in graph:
        neighborList = graph[gene2]
        neighborList[gene1] = pcc
    else:
        graph[gene2] = {gene1:pcc}

print(len(graph))

max = 0
maxGene = ""
for gene in graph:
      if (len(graph[gene]))>max:
        max = len(graph[gene])
        maxGene = gene

print("The gene with most connections is "+maxGene+" with "+str(max)+" connections.")

# now find the clustering coefficient of the gene with most neighbors

neighbors = graph[maxGene]
d = len(graph[maxGene]) # number of neighbors, same value as variable max here
e = 0 # for counting edges between the neighbors
for n in neighbors:
    for n2 in neighbors:
        if n!=n2 and n2 in graph[n]:
            e = e+1
# each edge is counted twice in the for loops above
# so divide by 2
e = e/2
cc = e/((d*(d-1))/2)
print("Clustering coefficient of "+maxGene+" is "+str(cc))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

First let's set up the fold-change increases in `x`:

```python
X = np.ones(600)

X[200:400] = 2
X[400:] = 4

plt.plot(X)
plt.xlabel("time")
plt.ylabel("X");
```

Now we integrate the responses in Y and Z for the IF-FFL:

$$\frac{dY}{dt} = \beta_1 X - \alpha_1 Y$$

$$\frac{dZ}{dt} = \frac{\beta_2 X}{Y} - \alpha_2 Z$$

```python
# parameters
beta1 = 10
beta2 = 12
alpha1 = 3
alpha2 = 4
dt = 0.1

# initilize Y and Z arrays to integrate
Y = np.zeros(len(X))
Z = np.zeros(len(X))
Y[0] = 1

# integration of Y and Z over time
for t in range(len(X)-1):
    dY = beta1*X[t] - alpha1*Y[t]
    Y[t+1] = Y[t] + dY*dt
    
    dZ = beta2*X[t]/Y[t] - alpha2*Z[t]
    Z[t+1] = Z[t] + dZ*dt  
    
t = np.arange(len(X))*dt 

# plotting
fig, ax = plt.subplots(3,1, figsize=(5,7))
ax[0].plot(t,X)
ax[0].set_ylabel("X")

ax[1].plot(t,Y)
ax[1].set_ylabel("Y")

ax[2].plot(t,Z)
ax[2].set_ylabel("Z")
ax[2].set_xlabel("time");
```

This does not display FCD behavior based on how Z is repressed. There is a large amount of initial Z that is slowly whittled away by each increase in Y. There is no stability within the amount of Z present at any point except for the very end of each step. 

```python
# parameters
beta1 = 10
beta2 = 12
alpha1 = 3
alpha2 = 4
dt = 0.1

# initilize Y and Z arrays to integrate
Y = np.zeros(len(X))
Z = np.zeros(len(X))
Y[0] = 1

# integration of Y and Z over time
for t in range(len(X)-1):
    dY = X[t] - Y[t]
    Y[t+1] = Y[t] + dY*dt
    
    dZ = X[t] - Y[t]*Z[t]
    Z[t+1] = Z[t] + dZ*dt  
    
t = np.arange(len(X))*dt 

# plotting
fig, ax = plt.subplots(3,1, figsize=(5,7))
ax[0].plot(t,X)
ax[0].set_ylabel("X")

ax[1].plot(t,Y)
ax[1].set_ylabel("Y")

ax[2].plot(t,Z)
ax[2].set_ylabel("Z")
ax[2].set_xlabel("time");
```

This set of equations does not show FCD behavior. For each repression of Z, it returns to the same steady-state value as its previous iteration, but the pulses are not the same height. This cannot be generalized into the mathematical form from the textbook, as the y in the second term fundamentally changes the setup of the Z production (and thus why all the pulses are different heights)

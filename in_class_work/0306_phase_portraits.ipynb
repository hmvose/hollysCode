## Phase portraits revisited!

Using the `phaseportrait` library, we will be able to make some pretty nice phase portraits in Python! 

Please reference the link below to learn more about how to use the `PhasePortrait2D` function:

https://phaseportrait.github.io/phaseportrait2d/#defining-range

For our first example, let's consider the following set of differential equations:

$$
\begin{align}
\frac{dx}{dt} &= \frac{1}{1+y} -x \\[1em] 
\frac{dy}{dt} &= x - y
\end{align}
$$

The first thing we need to do is define the differential equations themselves, in a function called `DEs` below. This function takes in $x$ and $y$ and returns the corresponding $dx/dt$ and $dy/dt$ values.

```python
def DEs(x,y):
    dx_dt = 1/(1+y) - x
    dy_dt = x - y
    return dx_dt, dy_dt
```

With our differential equations defined, we can now pass this function into the `PhasePortrait2D` function. Again, you might want to visit the developer's website to familiarize yourself with all the arguments for this function.

```python
portrait = pp.PhasePortrait2D(DEs, Range=[(-2, 1), (-2, 1)], 
                              xlabel="X", ylabel="Y", Title=" ", color="gray", MeshDim = 30)

x = np.linspace(-2,1,30)
y = np.linspace(-2,1,30)

null_x = 1/(1+y)
null_y = x

plt.plot(null_x, y, linewidth=3);
plt.plot(x, null_y, linewidth=3);
plt.legend(["null_x","null_y"])
portrait.plot();
```

## Exercise 2:

Repeat this analysis (nullclines and phase portrait) for our second set of differential equations:

$$
\begin{align}
\frac{dx}{dt} &= 1 - xy \\[1em] 
\frac{dy}{dt} &= x - y
\end{align}
$$

As a reminder, the fixed points occur at $x, y = 1$  and $x, y = -1$ for this system. So you'll want to make sure your plot includes these regions. How does a saddle point look in a phase diagram?

```python
def DE2s(x,y):
    dx_dt = 1 - x*y
    dy_dt = x - y
    return dx_dt, dy_dt
```

```python
portrait = pp.PhasePortrait2D(DE2s, Range=[(-2, 2), (-2, 2)], 
                              xlabel="X", ylabel="Y", Title=" ", color="gray", MeshDim = 30)

x = np.linspace(-2,2,30)
y = np.linspace(-2,2,30)

null_x = 1/y
null_y = x

plt.plot(null_x, y, linewidth=3);
plt.plot(x, null_y, linewidth=3);
plt.legend(["null_x","null_y"], loc="lower right")
portrait.plot();
```

## Exercise 3:

Lastly, repeat this analysis (nullclines and phase portrait) for our final set of differential equations:

$$
\begin{align}
\frac{dx}{dt} &= \frac{1}{y} - x \\[1em] 
\frac{dy}{dt} &= \frac{x^4}{1+x^4} - y
\end{align}
$$

As a reminder, the fixed point occurs at $\approx x = 1.3$ and $\approx y = 0.75$, so you'll want to make sure your plot includes this region. 

```python
def DE3s(x,y):
    dx_dt = 1/y - x
    dy_dt = x**4 / (1+x**4) - y
    return dx_dt, dy_dt
```

```python
portrait = pp.PhasePortrait2D(DE3s, Range=[(-0.5, 2), (-0.5, 2)], 
                              xlabel="X", ylabel="Y", Title=" ", color="gray", MeshDim = 30)

x = np.linspace(-0.5,2,30)
y = np.linspace(-0.5,2,30)

null_x = 1/y
null_y = x**4/(1+x**4)

plt.plot(null_x, y, linewidth=3);
plt.plot(x, null_y, linewidth=3);
plt.legend(["null_x","null_y"], loc="lower right")
portrait.plot();
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

$$Y_P = \frac{v(S)XY_T}{v(S)X + v_ZZ}$$

```python
v_s = np.linspace(0,3,100)
X = 5
Y_t = 2
v_z = 2
Z_vals = [10, 11, 12, 13]
point_compare = []


for i in Z_vals:
    Y_p = v_s*X*Y_t/(v_s*X + v_z*i)
    plt.plot(v_s,Y_p)
    
    point_compare.append(Y_p[49])
plt.legend(["Z=10","Z+10%","Z+20%","Z+30%"])
```

```python
for i in range(len(point_compare)-1):
    print(np.abs(point_compare[i+1]-point_compare[0]) / point_compare[0])
```

Each 10% increase in Z results in about a 6% decrease in Y_p, which is supported by the percentage differences calculated above.

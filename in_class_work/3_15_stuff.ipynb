```python
beta_1 = 10
alpha_1 = 5

beta_2 = 15
alpha_2 = 10

x = np.ones(100)

x[25:50] = 2
x[50:100] = 4
plt.plot(x)
```

```python
x = np.ones(100)

x[25:50] = 2
x[50:100] = 4
Y = np.zeros(100)
Y[0] = 1
Z = np.zeros(100)

for i in range(len(x)-1):
    dY = beta_1*x[i] - alpha_1*Y[i]
    Y[i+1] = Y[i] + dY*dt
    
    dZ = beta_2*x[i]/Y[i] - alpha_2*Z[i]
    Z[i+1] = Z[i] + dZ*dt
    
t = np.arange(len(x))*dt 
plt.plot(t,x)
plt.plot(t,Y)
plt.plot(t,Z)
plt.legend(["X Input","Y Production","Z Production"])
```
